We rely on the github repository Solar Wind Tools: https://github.com/08walkersj/Solar_Wind_Tools

The batch run function run_all incorporates everything that we need and use in this study.
- We set start and end year to 2000 and 2003 because this covers more than the period that has all the conjugat onset data points in our study.
- In the study we use a 2 hour window and apply a 20 minute shift when associating it with the onset dataset
- the run strings are the same as the default and perform all the available functions to retrieve the solar wind, magnetic index and dipole tilt dataset that is used in this study

```python
from Solar_Wind_Tools.run_all import run_all
run_all('./omni_data.hdf5',
        key='omni_window_2h_20', window=2*60, shift=20, start_year= 2000, 
        end_year=2003, run=['Download', 'Statistics', 'Coupling', 'Dipole', 'Time Shift'])
```
